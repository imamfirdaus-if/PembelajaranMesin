# Data Preparation




```python
#Menghubungkan source code dengan direktori pada google drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
import os, re, glob, cv2, numpy as np

path = os.path.abspath('QUIZ 11 Teori dan Praktikum Pembelajaran Mesin Imam Firdaus 1197050051 D.ipynb')
path = re.sub('[a-zA-Z\s._]+$', '', path)
dirs = os.listdir('/content/drive/My Drive/dataset/pneumonia/pneumonia-xray/')
label = 0
im_arr = []
lb_arr = []
X = []
y = []
for i in dirs:
    count = 0
    for pic in glob.glob('/content/drive/My Drive/dataset/pneumonia/pneumonia-xray/'+i+'/*'):
        im = cv2.imread(pic) #membuka gambar
        im = cv2.resize(im,(70,70))
        im = np.array(im) #merubah gambar menjadi array
        count = count + 1
        X.append(im)
        y.append(label)
        if(count <= 10): #Sample
            im_arr.append({str(i):im})
    print("Jumlah dataset xray "+str(i)+" : "+str(count)+" sample")
    label = label + 1
    lb_arr.append(i)
X = np.array(X)
y = np.array(y);
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(3, 7, figsize=(20, 10))
cnt = 0
row = 0
col = 0
for i in im_arr:
    for key, value in i.items():
        if(cnt==7):
            row = row + 1
            col = 0
            cnt = 0
        axs[row, col].imshow(value)
        axs[row, col].set_title(key)
        cnt = cnt + 1
        col = col + 1
plt.show()
```

# Modeling



```python
#Preprocessing
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

X_train = X_train.astype('float32') #konfigurasi tipe data x_train menjadi float32
X_test = X_test.astype('float32') #konfigurasi tioe data x_test menjadi float32
X_train /= 255 #merubah value x_train diantara 0 - 1
X_test /= 255 #merubah value x_test diantara 0 - 1
y_train = to_categorical(y_train, 2) #merubah label menjadi binary code
y_test = to_categorical(y_test, 2) #merubah label menjadi binary code
```

```python
#Arsitektur
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten, Dense
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3),activation='relu',input_shape=(70,70,3)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(32, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(2, activation='softmax'))
```

```python
#Compile
from tensorflow.keras.optimizers import SGD
epochs = 25
lrate = 0.01
decay = lrate/epochs
sgd = SGD(lr=lrate, momentum=0.9, decay=decay, nesterov=False)
model.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=['accuracy'])
print(model.summary())
```

```python
#Training
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=epochs, batch_size=32)
scores = model.evaluate(X_test, y_test, verbose=0)
print("Accuracy: %.2f%%" % (scores[1]*100))
```

```python
#Prediksi
y_pred = model.predict(X_test)
```

```python
#Iterasi X_test
y_prediction_n = []
y_test_n = []
for i in range(0, len(y_pred)):
    max_n = max(y_pred[i])
    max_m = max(y_test[i])
    index_n = [index_n for index_n, j in enumerate(y_pred[i]) if j == max_n]
    index_m = [index_m for index_m, j in enumerate(y_test[i]) if j == max_m]
    y_prediction_n.append(index_n)
    y_test_n.append(index_m)
```

# Evaluasi



```python
#Hasil Klasifikasi
from sklearn.metrics import classification_report
print(classification_report(y_test_n, y_prediction_n, target_names=["Pneumonia","Normal"]))
```
