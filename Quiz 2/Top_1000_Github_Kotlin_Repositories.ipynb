```python
df['id']=df['id'].fillna(0)
df['stars']=df['stars'].fillna(0)
df['forks']=df['forks'].fillna(0)
df['subscribers']=df['subscribers'].fillna(0)
df['open_issues']=df['open_issues'].fillna(0)
```

```python
df.corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

korelasi = df.corr()
plt.figure(figsize=(15, 10))
sns.heatmap(korelasi[(korelasi >= 0.5) | (korelasi <= -0.4)], cmap='viridis', vmax=1.0, vmin=-1.0, linewidths=0.1, 
            annot=True, annot_kws={"size": 8}, square=True).set_title('Korelasi Antar Variabel')
plt.show()
```

```python
#Perbandingan top 1000 user github berdasarkan jumlah subsribers dan jumlah stars
git_total = df.groupby("username")["subscribers"].sum().sort_values()
git_total.plot(kind="pie", fontsize=10, figsize=(8,40), title='Jumlah subscribers')
plt.show()

git_total = df.groupby("username")["stars"].sum().sort_values()
git_total.plot(kind="pie", fontsize=10, figsize=(8,40), title='Jumlah stars')
plt.show()
```

```python
sns.boxplot(df['stars'])
```

```python
# Diagram tebar Top 1000 Github Repositories berdasarkan subscribers dan stars
fig, ax = plt.subplots(figsize = (10,10))
ax.scatter(df['subscribers'], df['stars'])

# x-axis label
ax.set_xlabel('Total subscribers')
 
# y-axis label
ax.set_ylabel('Total stars')
plt.show()
```
